## SQL code
### for getting **netsuite data** (doubleclick to view the code in normal way)

SELECT
    t.TRANSACTION_ID,
    t.TRANID,
    t.TRANSACTION_TYPE,
    t.TRANDATE,
    t.EXCHANGE_RATE,
    t.CURRENCY_ID,
    t.ORDER_REF,
    t.MERCHANT_ACCOUNT,
    t.BATCH_NUMBER,
    t.IS_NON_POSTING,
    t.DATE_LAST_MODIFIED,
    tl.TRANSACTION_LINE_ID,
    tl.AMOUNT,
    tl.AMOUNT_FOREIGN,
    tl.SUBSIDIARY_ID,
    tl.ACCOUNT_ID,
    tl.COMPANY_ID,
    tl.DATE_LAST_MODIFIED
FROM
    netsuite.TRANSACTIONS AS t
JOIN
    netsuite.TRANSACTION_LINES AS tl
ON
    t.TRANSACTION_ID = tl.TRANSACTION_ID
WHERE
    t.BATCH_NUMBER = 138
    and t.MERCHANT_ACCOUNT = 'JetBrainsEUR'

```python
netsuite_data = pd.read_csv('netsuite_138.csv')
```

```python
netsuite_data
```

```python
merged_data = df.merge(netsuite_data, left_on=['ORDER_REF'], right_on=['ORDER_REF'], suffixes=('_csv', '_netsuite'))
```

```python
merged_data
```

```python
filtered_data = merged_data[(merged_data['AMOUNT_FOREIGN'] > 0) & (merged_data['ACCOUNT_ID'] == 2198)]
```

```python
filtered_data['GROSS_difference'] = filtered_data['GROSS'] - filtered_data['AMOUNT_FOREIGN']
```

```python
filtered_data.head()
```

```python
discrepancies = filtered_data[(filtered_data['GROSS_difference'] != 0)] 
```

```python
discrepancies.head()
```

```python
total_diff = discrepancies['GROSS_difference'].sum()
print(f"Total Diff Amount: {total_diff}")
```
